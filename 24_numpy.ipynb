## Difference in List and Arrays

1. A List cannot directly handle Mathematical operation while Array can .
2. An array consumes less memory than list .
3. Using array is faster than list .
4. A list can store multiple datatype while you cannot do it with arrays .
5. A list is easier to modify since a list store each element individually it is easier to add and delete an element than an array does
6. In list one can have nested data with different size , while you cannot do the same with array 

## Numpy - Creating Arrays , Slicing and Attributes

```python
# Creation of arrays 
arr = np.array([10,20,30,40])
print(arr)
print(type(arr))

arr2 = np.array([[10,20,30,40],[50,60,70,80]])
print(arr2)
print(type(arr2))
```

```python
#slicing 
arr = np.array([10,20,30,40])
print(arr[3:])
print(arr[2:])
print(arr[:3])

arr2 = np.array([[10,20,30,40],[50,60,70,80]])
print(arr2[0:2,0:3])
print(arr2[0,1:3]) #[20 30]
print(arr2[1,1:3]) #[60 70]
```

```python
#attributes of array
arr = np.array([[10,20,30,40],[50,60,70,80],[20,40,60,80]])
print(np.shape(arr)) #(3,4)
print(np.size(arr)) #12 ---gives elements 
print(np.ndim(arr)) #2
print(arr.dtype)
```

## Inspecting an array 

```python
a = [30,40,20,40,30]
b = [[20,10],[5,30]]
arr = np.array(a)
arr2 = np.array(b)
print(arr.shape) #(5,)
print(arr2.shape) #(2,2)
print(len(arr)) #5
print(len(arr2)) #2
print(np.size(arr)) #5
print(type(arr))
print(arr.dtype)
print(arr.astype(float))
print(arr.dtype)
```

## Mathematical Operations and Functions on Arrays

```python
arr1 = np.array([30,40,60,20,10])
arr2 = np.array([30,20,10,20,20])
print(arr1+arr2)
print(np.add(arr1,arr2))
arr3 = np.array([[30,40],[60,20]])
arr4 = np.array([[30,20],[10,20]])
print(arr3+arr4)
```

```python
arr1 = np.array([30,40,60,20,10])
arr2 = np.array([30,20,10,20,20])
print(np.subtract(arr1,arr2))
print(arr1-arr2)

# Same with multiply and divide 
```

```python
arr1 = np.array([3,4,6,2,10])
arr2 = np.array([3])
np.power(arr1,arr2)
```

```python
arr1 = np.array(([  27,   64,  216,    8, 1000]))
np.sqrt(arr1)
```

## Combining and splitting arrays 

```python
a = [30,40,50]
b = [5,5,3]
print(a+b)
```

```python
#concatenate
import numpy as np
arr1 = np.array([30,40,50])
arr2 = np.array([5,5,3])
np.concatenate([arr1,arr2])
```

```python
arr1 = np.array([[30,40],[50,40]])
arr2 = np.array([[5,5],[7,8]])
np.concatenate([arr1,arr2],axis=1)
```

```python
#horizontal stack ---->horizontal concatenation
print(np.hstack([arr1,arr2]))
```

```python
#vertical stack ------>vertical concatenation
print(np.vstack([arr1,arr2]))
```

```python
#split arrays 
a = np.array([20,40,30,40,10,20])
print(np.array_split(a,3))
print(a)
```

## Adding and removing element in Arrays

```python
#append
a = np.array([20,40,60,80])
print(np.append(a,90))

b = np.array([[20,40],[60,80]])
np.append(b,90)
np.append(b,[3,4])
```

```python
#insert
a = np.array([20,40,60,80])
np.insert(a,1,5) #array , index,value. 
# array([20,  5, 40, 60, 80])

b = np.array([[20,40],[60,80]])
np.insert(b,1,[50,60],axis=1)
# array([[20, 50, 40],
#    [60, 60, 80]])
np.insert(b,1,[50,60],axis=0)
# array([[20, 40],
#    [50, 60],
#    [60, 80]])
np.insert(b,[1,2],[50],axis=0)
# array([[20, 40],
    # [50, 50],
    # [60, 80],
    # [50, 50]])
    
```

```python
#delete 
a=  np.array([20,40,60,80])
np.delete(a,1)
# array([20, 60, 80])
```

## Numpy - Sort Filter and Search

```python
ar = np.array([7,6,9,2,8])
np.sort(ar)
ar2 = np.array([[7,6,9],[4,2,5]])
np.sort(ar2)
```

```python
#search
ar = np.array([7,6,9,2,8])
s = np.where(ar==6)
print(s)
evenElementIndex = np.where(ar%2==0)
print(evenElementIndex)
```

```python
#search sorted
ar = np.array([1,2,3,4,5])
ss = np.searchsorted(ar,4)
ss
```

```python
#filter
ar = np.array([20,30,40,50])
fa = [True,False,True,False]

new = ar[fa]
new
```

```python
ar = np.array([20,33,43,50])
fa = ar>35
new = ar[fa]
new
evenfa = ar%2==0
neweven = ar[evenfa]
neweven
```

## Aggregating function in Numpy

```python
a = np.array([20,40,50,70])
print(np.sum(a))
print(np.max(a))
print(np.min(a))
print(np.size(a))
print(np.mean(a))
print(np.cumsum(a)) #cummilative sum 
print(np.cumprod(a))
```

```python
a = [100,150,199,200,250,130]
b = [10,50,30,40,30,10]

price = np.array(a)
quantity = np.array(b)
print(price,"\n",quantity)

c=np.cumprod([price,quantity],axis = 0)
print(c[1].sum())
```

## Statistical Functions of Numpy

```python
import numpy as np
import statistics as stats
baked_food = [200,300,150,130,200,280,170,188]

a = np.array(baked_food)
print(np.mean(baked_food)) #sum of all values/number of values
print(np.median(baked_food)) #central value after sorting
print(stats.mode(baked_food))
print(np.std(baked_food))
print(np.var(baked_food))
```

```python
#-1 inversely relationship 
# 1 proportional relatsionship 
# 0 no relationship

tabacoo_consumption = [30,50,10,30,50,40]
deaths = [100,120,70,100,120,112]

print(np.corrcoef([tabacoo_consumption,deaths]))
```

```python
price = [300,100,350,150,200]
sales = [10,20,7,17,3]

print(np.corrcoef([price,sales]))
```
